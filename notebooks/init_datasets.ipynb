```python
target_categories = {
    # obesity category
    "obesity_I_II_III": [
        "Obesity_Type_I", "Obesity_Type_II", "Obesity_Type_III"
    ],

    # severe obesity category
    "severe_obesity_II_III": [
        "Obesity_Type_II", "Obesity_Type_III"
    ],

    # normal weight individuals
    "normal_weight": [
        "Normal_Weight"
    ],

    # type II obesity individuals
    "type_II_obesity": [
        "Obesity_Type_II"
    ],

    # type III obesity individuals
    "type_III_obesity": [
        "Obesity_Type_III"
    ],

    # underweight individuals
    "insufficient_weight": [
        "Insufficient_Weight"
    ],

    # overweight level II
    "level_II_overweight": [
        "Overweight_Level_II"
    ]
}
```

Calculating `Statistic Parity Difference` and `Disparate Impact` in obesity data frame

Both are part of the fairness metrics suite

```python
def fairness_test(df, sensitive_attribute, target_col, positive_classes):
    y_bin = df[target_col].isin(positive_classes).astype(int)

    rates = y_bin.groupby(df[sensitive_attribute]).mean()

    p_female = rates.get("Female", float("nan"))
    p_male = rates.get("Male", float("nan"))

    spd = p_female - p_male # Statistic Parity Difference
    di = p_female - p_male if p_male > 0 else float("inf") # Disparate Impact
    
    return p_female, p_male, spd, di
```

```python
results = []
for name, positives in target_categories.items():
    p_f, p_m, spd, di = fairness_test(df, 
                                      sensitive_attribute="gender", 
                                      target_col="target", 
                                      positive_classes=positives
                                    )
    results.append({
        "task": name,
        "positives": ", ".join(positives),
        "P(Y=1|Female)": p_f,
        "P(Y=1|Male)": p_m,
        "SPD (F - M)": spd,
        "DI (F / M)": di,
    })

fairness_screen = pd.DataFrame(results)
fairness_screen
```
